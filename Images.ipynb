```python
input_image=Image.open("1.jpg")
pixel_map=input_image.load()
w,h=input_image.size
for i in range(w//2):
    for j in range(h):
        r,g,b=input_image.getpixel((i,j))
        gs=(0.299*r+0.587*g+0.114*b)
        pixel_map[i,j]=(int(gs),int(gs),int(gs))

input_image.save("test.png")
```
